```python
get_dataframe = pd.DataFrame(columns=['행정동명','상권업종중분류명','위도','경도'])

for idx in range(len(list(df.상권업종중분류명))):
    if df.상권업종중분류명[idx] in use:
        get_dataframe.loc[idx] = list(df.loc[idx])
        # break

get_dataframe.head()
```

```python
get_dataframe.to_csv('../data/total_data/부산상권정보_handling.csv',encoding='cp949')
```

```python
dong = {}

for idx in range(len(get_dataframe)):
    com = get_dataframe.iloc[idx]['행정동명']
    if com in dong.keys():
        dong[com] += 1
    else:
        dong[com] = 1
    
dong
```

```python
len(dong.keys()), len(dong.values())
```

```python
dong_dataframe = pd.DataFrame({'동 이름':dong.keys(),'업장 수':dong.values()})
dong_dataframe
```
