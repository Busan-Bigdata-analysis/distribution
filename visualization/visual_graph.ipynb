```python
dong = {}

for idx in range(len(get_dataframe)):
    com = get_dataframe.iloc[idx]['행정동명']
    if com in dong.keys():
        dong[com] += 1
    else:
        dong[com] = 1
    
dong
```

```python
empty_place_df = pd.DataFrame(columns=['상가번호', '역사명', '상가명', '업종', '면적(㎡)', '연간계약금액(부가세포함)', '계약시작일', '계약종료일','비고'])
place_num = len(place_df)

for idx in range(place_num):
    if bool(list(place_df['계약시작일'])[idx]==0):
        empty_place_df.loc[idx] = list(place_df.iloc[idx])
```

```python
empty_place_df.reset_index(drop=True, inplace=True)
```

```python
empty_place_df.head(5)
```

```python
list(empty_place_df['역사명'])[0]
```

```python
station_dic = {}

for idx in range(len(empty_place_df)):
    station_name = list(empty_place_df['역사명'])[idx]
    if station_name in station_dic.keys():
        station_dic[station_name] += 1
    else:
        station_dic[station_name] = 1
```
