```python
import sys
import pandas as pd
import os
```

```python
df = pd.read_csv('../data/total_data/통합_사용_데이터_추가수정_handling.csv',encoding='cp949',index_col=0)
df.head()
```

```python
df = df[['역사명', '동별', '역위도', '역경도',
       '총 상가 수', '공실 상가 수', '전체 상가 면적', '공실 상가 면적', '일평균승하차 레벨',
       '업장수 레벨', '1인 가구수 레벨', '2인 가구수 레벨', '20~40대 인구 수 레벨', '고속버스 터미널', '환승', '기차역',
       '여객 터미널', '주변 관광지','주변 대학교']]
df.head()
```

```python
df = df.sort_values(by='전체 상가 면적',ascending=False).iloc[:50]
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.columns
```

```python
rank_df = df[['역사명','일평균승하차 레벨', '업장수 레벨', '1인 가구수 레벨', '2인 가구수 레벨', '20~40대 인구 수 레벨','고속버스 터미널', '환승', '기차역', '여객 터미널', '주변 관광지','주변 대학교']].fillna(0)
rank_df.head()
```

```python
rank_df['종합점수'] = 0

for idx,line in enumerate(rank_df.iloc):
    # 기준 열의 합
    main_score = sum(line[1:6])
    # 기준 열의 갯수
    div = len(line[1:6])

    for item in line[6:]:
        if item:
            main_score += 10

    rank_df['종합점수'].iloc[idx] = main_score/div
```

```python
rank_df.sort_values(by='종합점수',ascending=False).reset_index(drop=True)
```

```python
total = df[['역사명', '동별', '역위도', '역경도', '총 상가 수', '공실 상가 수', '전체 상가 면적', '공실 상가 면적']].merge(rank_df,on=['역사명']).sort_values(by='종합점수',ascending=False).reset_index(drop=True)
total.head()
```

```python
total.to_csv('../data/total_data/종합_점수_계산_가중치_x.csv',encoding='cp949')
```
