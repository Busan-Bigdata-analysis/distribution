```python
import pandas as pd
```

```python
people_data = pd.read_csv('../data/읍면동별_연령_각세__및_성별_한국인_현황__전체.csv',encoding='cp949',skiprows=1)
people_data.head()
```

```python
people_data = people_data.loc[20:40]
people_data
```

```python
people_data.drop(columns=['연령별(2)'],inplace=True)
people_data.head()
```

```python
age_dic = {}

for ages in people_data:
    age_dic[ages] = sum(people_data[ages])

print(len(age_dic.keys()))
```

```python
age_data = pd.DataFrame({'동별':age_dic.keys(), '20~40대 인구 수':age_dic.values()})
age_data.head()
```

```python
age_data.to_csv('../data/total_data/타겟층_인구_수.csv',encoding='cp949')
```
