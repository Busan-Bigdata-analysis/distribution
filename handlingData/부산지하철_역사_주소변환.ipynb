데이터 출처 : https://www.data.go.kr/data/15041101/fileData.do

```python
import pandas as pd
```

```python
# 데이터 불러오기
df1 = pd.read_csv("./국가철도공단_부산_지하철_주소데이터_20211123.csv", encoding="cp949")
df1
```

```python
# 결측치 제거
df1 = df1.dropna(axis=0)
df1
```

```python
#필요한 열만 추출
df1.drop(["철도운영기관명","도로명주소"],axis=1, inplace=True)
df1
```

```python
# 동, 리만 추출
df1["행정동"] = df1["지번주소"].str.split().str[2:3].astype("str")
df1
```

```python
df1["행정동"] = df1["행정동"].str.split("[").str.get(-1).str.replace("]","")
df1["행정동"] = df1["행정동"].str.split("'").str.get(1)
```

```python
df1
```

```python
# 기장군은 면,리
df1.loc[112, "행정동"] = "철마면 고촌리"
df1.loc[113, "행정동"] = "철마면 안평리"

# 양산
df1.loc[78, "행정동"] = "동면 가산리"
df1.loc[79, "행정동"] = "물금읍 증산리"
df1.loc[80, "행정동"] = "물금읍 범어리"
df1.loc[81, "행정동"] = "동면 석산리"
df1.loc[82, "행정동"] = "중부동"

# "*가" 제거
df1.loc[12, "행정동"] = "서대신동"
df1.loc[13, "행정동"] = "동대신동"
df1.loc[14, "행정동"] = "토성동"
df1.loc[15, "행정동"] = "남포동"
df1.loc[16, "행정동"] = "남포동"
df1.loc[17,"행정동"] = "중앙동"
```

```python
df1.head(83)
```

```python
df1.drop(["지번주소"],axis=1, inplace=True)
df1
```

```python
# 저장
df1.to_csv("부산지하철 역사별 행정동.csv", encoding="cp949", index=False)
```
