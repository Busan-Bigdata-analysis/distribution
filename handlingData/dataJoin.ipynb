```python
def readfile(file_name):
    f_type = file_name.split('.')[-1]
    if f_type == 'xlsx':
        return pd.read_excel(f'../data/total_data/{file_name}',index_col=0)
    elif f_type == 'csv':
        return pd.read_csv(f'../data/total_data/{file_name}',encoding='cp949',index_col=0)
    else:
        return None

readfile(dir_list[1]).head()
```

##### 지하상가 전체 갯수와 비어있는 상가, 사용중인 상가 갯수를 나타내기 위한 데이터 프레임

```python
# nan값을 0으로 대체하여 공실 여부 판단
station_store.fillna(0,inplace=True)
```

```python
store_info = {}

for idx in range(len(station_store)):
    item = station_store.iloc[idx]
    if list(item)[1] in store_info.keys():
        store_info[list(item)[1]]['상가 갯수'] += 1
        if list(item)[6] != 0.0:
            store_info[list(item)[1]]['사용중인 상가'] += 1
        else:
            store_info[list(item)[1]]['비어있는 상가'] += 1
    else:
        store_info[list(item)[1]] = {'상가 갯수':1,'사용중인 상가':0,'비어있는 상가':0}
        if list(item)[6] != 0.0:
            store_info[list(item)[1]]['사용중인 상가'] += 1
        else:
            store_info[list(item)[1]]['비어있는 상가'] += 1
```

```python
len(store_info.keys())
```

```python
i = 0
for item in store_info.keys():
    station_store_num.loc[i] = [item[:-1]]+list(store_info[item].values())
    i+=1
```

```python
tour_area = pd.read_csv('../data/total_data/2020 관광지별 시도별 관광객 현황(주변역 추가).csv')[['관광지명','위도','경도']]
tour_area.head()
```

```python
sub_area = pd.read_csv('../data/total_data/통합_사용_데이터_추가수정_handling.csv',encoding='cp949',index_col=0)[['역사명','역위도','역경도']]
sub_area['주변 관광지'] = 0
sub_area.head()
```

```python
from haversine import haversine
```

- 관광지와 역 사이의 거리 1.2 km

```python
for idx,sub_line in enumerate(sub_area.iloc):
    start = (sub_line[1],sub_line[2])
    for tour_line in tour_area.iloc:
        end = (tour_line[1],tour_line[2])
        length = haversine(start,end)
        if 1.2 > length:
            sub_area['주변 관광지'].iloc[idx] += 1
```
