# 부산광역시 동별 인구현황 전처리

# 코드 정의서
- a_pop : 행정동별 나이 인구수
- gu_gun(부산 행정구) : 강서구, 사상구, 사하구, 북구, 금정구, 동래구, 연제구, 부산진구, 서구, 동구, 남구, 중구, 영도구, 수영구, 해운대구, 기장군


```python
import pandas as pd
```

```python
# 데이터 불러오기
population_df = pd.read_csv("./data/부산광역시_읍면동별_연령_및_성별_202207.csv", encoding="cp949", index_col=False)
population_df
```

```python
# 열 제거
population_df.drop(columns="연령별(1)", inplace=True)
```

```python
# 행,열 변경
population_df = population_df.transpose()
```

```python
population_df.rename(columns=population_df.iloc[0], inplace=True)
```

```python
population_df.drop(population_df.index[0], inplace=True)
```

```python
population_df.reset_index(inplace=True)
```

```python
# 열 이름 변경
population_df.rename(columns={"연령별(2)":"성별", "소계":"합계","index":"행정구"}, inplace=True)
```

```python
# 행정구역별 나이 인구수 Dataframe 생성
a_pop = population_df.drop(columns="성별")
```

```python
# .을 문자로 인식하여 .1, .2포함 행 삭제
a_pop = a_pop[~a_pop['행정구'].str.contains("[.]")]
```

```python
a_pop.reset_index(inplace=True)
a_pop.drop(columns="index",inplace=True)
a_pop
```

```python
a_pop.to_csv("부산광역시 행정동.csv", index = False, encoding="cp949")
# 이후 엑셀로 수정
```
